```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split,GridSearchCV
from sklearn.model_selection import KFold
from sklearn.neighbors import KNeighborsRegressor
from sklearn import metrics
from statsmodels.stats.outliers_influence import variance_inflation_factor
```

```python
nba=pd.read_csv('nba_2013.csv')
nba
```

```python
nba.describe()
```

```python
nba.isnull().sum()
```

```python
nba.dtypes
```

```python
nba['fg.']=nba['fg.'].fillna(nba['fg.'].mean())
nba['x3p.']=nba['x3p.'].fillna(nba['x3p.'].mean())
nba['x2p.']=nba['x2p.'].fillna(nba['x2p.'].mean())
nba['efg.']=nba['efg.'].fillna(nba['efg.'].mean())
nba['ft.']=nba['ft.'].fillna(nba['ft.'].mean())
```

```python
nba.isnull().sum()
```

```python
nba=nba.drop(columns=['player','bref_team_id','season','pos'])
nba.columns
```

```python
X=nba[['age', 'g', 'gs', 'mp', 'fg', 'fga', 'fg.', 'x3p', 'x3pa', 'x3p.',
       'x2p', 'x2pa', 'x2p.', 'efg.', 'ft', 'fta', 'ft.', 'orb', 'drb', 'trb',
       'ast', 'stl', 'blk', 'tov', 'pf', 'season_end']]
Y=nba['pts']
print(Y)
```

```python
sc=StandardScaler()
X=sc.fit_transform(X)
```

```python
x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size=0.3, random_state=42)
```

```python
knn=KNeighborsRegressor()
knn.fit(x_train,y_train)
```

```python
y_pred=knn.predict(x_test)
print ("Regression score is:",format(metrics.r2_score(y_test, y_pred),'.4f'))
```

```python
param_grid = { 'algorithm' : ['ball_tree', 'kd_tree', 'brute'],
               'leaf_size' : [18,20,25,27,30,32,34],
               'n_neighbors' : [3,5,7,9,10,11,12,13]
              }
```

```python
gridsearch = GridSearchCV(knn, param_grid,verbose=3)
```

```python
gridsearch.fit(x_train,y_train)
```

```python
gridsearch.best_params_
```

```python
knn=KNeighborsRegressor(algorithm='ball_tree', leaf_size=18, n_neighbors=7)
knn.fit(x_train,y_train)
```

```python
y_pred=knn.predict(x_test)
print ("Regression score is:",format(metrics.r2_score(y_test, y_pred),'.4f'))
```
